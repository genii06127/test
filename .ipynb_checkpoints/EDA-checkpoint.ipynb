```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# data extraction

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
tips = sns.load_dataset("tips")    # 팁 데이터
flights = sns.load_dataset("flights")    # 여객운송 데이터
```

# 1. rugplot

```python
x = iris.petal_length.values
sns.rugplot(x)
plt.title('Iris Rug dPlot')
plt.show()
```

# 2. kdeplot

```python
sns.kdeplot(x)
plt.title('Iris kenel density plot')
plt.show()
```

# 3. distplot

```python
sns.distplot(x, kde = True, rug = True)
plt.title('iris dist plot')
plt.show()
```

# countplot

```python
sns.countplot(x= 'class', data = titanic)
plt.title('titanic class customer')
plt.show()
```

# jointplot

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris)
plt.title('joint plot')
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="kde")
plt.title("Joint Plot with Kernel Density Plot")
plt.show()
```

# 다차원 실수형 데이터

```python
sns.pairplot(iris)
plt.title('iris data pair plot')
plt.show()
```

# pairplot : separation with category

```python
sns.pairplot(iris, hue='species', markers=['o','s','D'])
plt.title('iris pair plot with hue')
plt.show()
```

```python

sns.heatmap(titanic.corr(), annot=True, fmt=".2f")
plt.show()
```
